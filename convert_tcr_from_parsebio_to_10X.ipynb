```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
bc = pd.read_csv('barcode_report.tsv', sep='\t', index_col=0)
clo_fre = pd.read_csv('clonotype_frequency.tsv', sep='\t', index_col=0)
tcr_anno = pd.read_csv('tcr_annotation_airr.tsv', sep='\t', index_col=0)
```

```python
final_tcr_file = pd.DataFrame(index=tcr_anno['cell_barcode'].values)
```

```python
final_tcr_file['is_cell'] = 'TRUE'

all_contig_ids = []
bc_dct = dict()
for i in final_tcr_file.index.values:
    if i not in bc_dct:
        contig_id = i + '_contig_1'
        bc_dct[i] = 1
    else:
        bc_dct[i] = bc_dct[i] + 1
        contig_id = i + '_contig_{}'.format(bc_dct[i])
    all_contig_ids.append(contig_id)
final_tcr_file['contig_id'] = all_contig_ids

final_tcr_file['high_confidence'] = 'TRUE'
final_tcr_file['length'] = 0
final_tcr_file['chain'] = tcr_anno['locus'].values
final_tcr_file['v_gene'] = tcr_anno['v_call'].values
final_tcr_file['d_gene'] = tcr_anno['d_call'].values
final_tcr_file['j_gene'] = tcr_anno['j_call'].values
final_tcr_file['c_gene'] = tcr_anno['c_call'].values
final_tcr_file['full_length'] = 'TRUE'
final_tcr_file['productive'] = ['TRUE' if i == 'T' else False for i in tcr_anno['productive'].values]
final_tcr_file['fwr1'] = 'Unassigned'
final_tcr_file['fwr1_nt'] = 'Unassigned'
final_tcr_file['cdr1'] = 'Unassigned'
final_tcr_file['cdr1_nt'] = tcr_anno['cdr1'].values
final_tcr_file['fwr2'] = 'Unassigned'
final_tcr_file['fwr2_nt'] = 'Unassigned'
final_tcr_file['cdr2'] = 'Unassigned'
final_tcr_file['cdr2_nt'] = tcr_anno['cdr2'].values
final_tcr_file['fwr3'] = 'Unassigned'
final_tcr_file['fwr3_nt'] = 'Unassigned'
final_tcr_file['cdr3'] = tcr_anno['cdr3_aa'].values
final_tcr_file['cdr3_nt'] = tcr_anno['cdr3'].values
final_tcr_file['fwr4'] = 'Unassigned'
final_tcr_file['fwr4_nt'] = 'Unassigned'
final_tcr_file['reads'] = tcr_anno['read_count'].values
final_tcr_file['umis'] = tcr_anno['transcript_count'].values

clono_dct = {}
for i in range(len(clo_fre)):
    clono_dct['{}<<>>{}'.format(clo_fre.index[i], clo_fre['TRB'].values[i])] = clo_fre['clonotype_id'].values[i]
bc['test'] = [
    '{}<<>>{}'.format(bc['TRA_cdr3_aa'][i], bc['TRB_cdr3_aa'][i])
    for i in range(len(bc))
]
bc['clonotype_id'] = [clono_dct[i] if i in clono_dct else 'Unassigned' for i in bc['test'].values]
bc_2_clonotype = dict(zip(bc.index.values, bc['clonotype_id'].values))

final_tcr_file['raw_clonotype_id'] = [
    bc_2_clonotype[i] for i in final_tcr_file.index.values
]

final_tcr_file['raw_consensus_id'] = 'Unassigned'
final_tcr_file['exact_subclonotype_id'] = 'Unassigned'
```

```python
final_tcr_file.to_csv('converted_tcr_file.csv')
```
